```python
# Plot budget vs gross using seaborn
# Scatterplot with budget vs gross
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color': 'red'}, line_kws = {'color': 'blue'}).set(title = 'Budget vs Gross Earnings')
```

```python
correlation_matrix = df.corr(method = 'pearson')
# Pearson, kendall, spearman correlations. In this case, we are using Pearson.

# Pearson measures linear r/s between 2 cont. variables. Assume that they are normally distributed and have a linear r/s. Used when analysing the strength and direction of a linear r/s between 2 quantitative variables. Sensitive to outliers.
# Kendall measures the strength of dependence between two variables, regardless of the functional form of the r/s. Assume that variables are ordinal or continuous. Used when analysing the association between 2 variables that may not have a linear r/s or when dealing with ordinal data.
# Spearman measures the strength and direction of monotonic r/s between 2 variables. Assume that variables are ordinal or continuous. Used when analysing the strength and direction of a monotonic r/s (i.e. 1 variable consistently increases or decreases as the other variable does.)
```

```python
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
